# Python Script for Showing tweets from JSON file in Tabluar Form

```python
import numpy as np                                                    
import pandas as pd                                                   
import json                                                         
```

Function for opening json file

```python
def json_input(location):
    with open(location) as json_file:  
        data = json.load(json_file)
        return data
    
```

```python
data = json_input("tweets.json")
```

Converting Json data to DataFrame

```python
tweets = pd.DataFrame.from_dict(data)
tweets
```
